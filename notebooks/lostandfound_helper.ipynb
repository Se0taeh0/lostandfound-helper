````python
import json
import re

def extract_slots(user_input):
    prompt = f"""다음 문장에서 물건종류, 색상/특징, 장소, 날짜를 추출해줘.
순수 JSON 텍스트만 출력해.

문장: "{user_input}"

출력 형식:
{{"item": "", "color": "", "location": "", "date": ""}}
"""
    message = client.messages.create(
        model="claude-haiku-4-5",
        max_tokens=200,
        messages=[{"role": "user", "content": prompt}]
    )
    raw_text = message.content[0].text

    # 코드블록 제거
    cleaned = re.sub(r'```json\s*|\s*```', '', raw_text).strip()

    return json.loads(cleaned)  # 이제 진짜 dict 형태로 반환됨

result = extract_slots("5월 중순쯤 라지브 초크역에서 파란색 아이폰을 잃어버렸어요")
print(result)
print(type(result))  # <class 'dict'> 나와야 정상
print(result['item'])  # '아이폰' 나와야 정상
````
